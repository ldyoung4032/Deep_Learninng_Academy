```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd
seed=0
np.random.seed(seed)
tf.random.set_seed(seed)

data = np.loadtxt("ThoraricSurgery.csv", delimiter=",")

X = data[:,0:17]
Y = data[:,17]

model = Sequential()
model.add(Dense(30,input_dim = 17, activation = 'sigmoid')) ## 은닉층, 노드갯수 30개, 입력변수 17개
model.add(Dense(1,activation = 'sigmoid')) ## 출력층, 노드갯수 1개

model.compile(loss='mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, Y, epochs=30, batch_size=10)
print("\n Accuracy : %.4f" % (model.evaluate(X, Y)[1]))

## tf와 달리 학습률이 존재하지 않는다
## optimizer = 'adam'안에 존재한다 디폴트:0.01

## SGD, RMSprop, Adagrad 등등
```

```python
# from keras import models
# from keras import optimizers

## compile은 모델의 파라미터 등 옵션을 정해주는 부분
## fit은 모델을 실제로 수행하는 부분

## keras는 tf를 자동화한 라이브러리
## 한 눈에 보기 쉽게 정리됨

## batch_size : 1명씩이 아니라 10명씩 넣어서 47번씩 나눠서 돌려보라는 것
## 너무 크면 학습 속도가 느려짐, 너무 작으면 실행값의 편차가 생겨서 결괏값이 불안정해질수 있음
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

## relu, sigmoid

training_data = [[0,0], [0,1], [1,0], [1,1]]
target_data = [[0], [1], [1], [0]]

model = Sequential()
model.add(Dense(32,input_dim = 2, activation = 'relu')) 
model.add(Dense(1,activation = 'sigmoid'))

model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics = ['binary_accuracy'])
model.fit(training_data, target_data, epochs=1000, verbose=2)
print(model.predict(training_data))
print("\n Accuracy : %.4f" % (model.evaluate(training_data, target_data)[1]))
```
