```python
soup = bs(resp.text)
soup
```

```python
header = soup.find('header')
body = soup.find('body')
```

```python
result_code = header.find('resultcode').text
result_msg = header.find('resultmsg').text
result_code, result_msg
```

```python
num_of_rows = body.find('numofrows').text
page_no = body.find('pageno').text
total_count = body.find('totalcount').text
num_of_rows, page_no, total_count
```

```python
items = body.find_all('item')

result_list = []
for item in items :
    result_list.append({
        'anldt' : item.find('anldt').text,
        'anlrst' : item.find('anlrst').text,
        'anlval' : item.find('anlval').text,
        'dv' : item.find('dv').text,
        'hjdcd' : item.find('hjdcd').text
    })
    
result_list
```

# 네이버 POST 요청(데이터랩)

```python
url = 'https://openapi.naver.com/v1/datalab/search'
headers = {
    "X-Naver-Client-Id": "KAgv_PdX089EimNALWw5",
    "X-Naver-Client-Secret": "AiuXZcQcAK",
    "Content-Type" : "application/json"
}

params = {
    'startDate' : '2019-08-03',
    'endDate' : '2020-08-03',
    'timeUnit' : 'date',
    'keywordGroups' : [
        {
            'groupName' :'코로나',
            'keywords' : ['코로나 현황', '코로나', '코로나 확산']
        },
        {
            'groupName' :'마스크',
            'keywords' : ['마스크', '공적 마스크', 'kf94', 'kf80']
        }
    ]
}
```

```python
type(params) ## dict를 json으로 바꿔주기

## 만약 xml로 만들어야 할 경우는 dicttoxml 라이브러리를 설치하여 사용
```

```python
import json
```

```python
data = json.dumps(params)
data
```

```python
resp = re.post(url, headers = headers, data=data) ## 리퀘 = 헤더 + 바디 + url
resp
```
