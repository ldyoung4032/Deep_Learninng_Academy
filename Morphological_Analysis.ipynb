## 제목, 내용

```python
import requests as re
from bs4 import BeautifulSoup as bs

resp = re.request("GET", "https://www.forbes.com/sites/adrianbridgwater/2019/04/15/what-drove-the-ai-renaissance/#6d947cec1f25")
soup = bs(resp.text)
print(resp.text)
```

```python
content = []
div_find = soup.find('div', class_ = 'article-body fs-article fs-responsive-text current-article')
p_find = div_find.find_all('p')
for p in p_find :
    content.append(p.text)
content = " ".join(content)
print(content)
```

### word_tokenize : 구둣점 기반

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
token1 = word_tokenize(content)
print(token1[:20])

# 'present-day'
```

### WordPunctTokenizer : 알파벳이 아닌 문자를 구분하여 토큰화

```python
from nltk.tokenize import WordPunctTokenizer
token2 = WordPunctTokenizer().tokenize(content)
print(token2[:20])

# 'present', '-', 'day'
```

### TreebankWordTokenizer : 정규표현식에 기반한 토큰화

```python
from nltk.tokenize import TreebankWordTokenizer
token = TreebankWordTokenizer().tokenize(content)
print(token[:20])
```

## 텍스트 전처리

```python
text = '인생은 모두가 함께하는 여행이다. 매일매일 사는 동안 우리가 할 수 있는 건 최선을 다해 이 멋진 여행을 만끽하는 것이다.'
print(text.split(' '))
```

```python
## 코모란 : 상세한 분석 가능 ('하', 'ㄹ')
from konlpy.tag import Komoran
komoran =Komoran()
komoran_tokens = komoran.morphs(text)
print(komoran_tokens)
```

```python
## 한나눔
from konlpy.tag import Hannanum
hannanum = Hannanum()
hannanum_tokens = hannanum.morphs(text)
print(hannanum_tokens)
```

```python
from konlpy.tag import Okt
okt = Okt()
okt_tokens = okt.morphs(text)
print(okt_tokens)
```

```python
from konlpy.tag import Kkma
kkma = Kkma()
kkma_tokens = kkma.morphs(text)
print(kkma_tokens)
```

## 품사 부착

```python
## 코모란
komoranTag = []
for token in komoran_tokens :
    komoranTag += komoran.pos(token)
print(komoranTag)
```

```python
## 한나눔
hannanumTag = []
for token in hannanum_tokens :
    hannanumTag += hannanum.pos(token)
print(hannanumTag)
```

```python
## Okt
oktTag = []
for token in okt_tokens :
    oktTag += okt.pos(token)
print(oktTag)
```

```python
## Kkma
kkmaTag = []
for token in kkma_tokens :
    kkmaTag += kkma.pos(token)
print(kkmaTag)
```

## 불용어 처리(Stopword)
자연어 처리를 위해 불필여한 요소를 제거하는 작업

불필요한 품사를 제거하는 작업과 불필요한 단어를 제거하는 작업으로 구성

불필요한 토큰을 제거함으로써 연산의 효율성을 높임

```python
from collections import Counter
Counter(oktTag).most_common()
```

```python
stopPos = ['Determiner', 'Adverb', 'Conjunction', 'Josa', 'PreEomi', 'Eomi', 'Suffix', 'Punctuation', 'Foreign', 'Alpha','Number', 'Unknown']

stopWord = []
#  stopWord = ['을', '는', '기']
word = []
for tag in oktTag :
    if tag[1] not in stopPos :
        if tag[0] not in stopWord :
            word.append(tag[0])

print(word)
```
