```python
## 생각 
## 1. for 5개 페이지 들어가기

def page_crawling(page_num, topic) :

    result_list = []

    for page in range(page_num) : ## post, query, page
        resp = re.request("GET", "https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=&post_blogurl_without=&query={}&sm=tab_pge&srchby=all&st=sim&where=post&start={}".format(topic,int(str(page)+"1")))
        soup = bs(resp.text)
        blog_find = soup.find('div', class_ = 'blog section _blogBase _prs_blg')
        blog = blog_find.find('ul', class_ = 'type01')
        blog_li = blog.find_all('li')

        for li in blog_li :
            find_a = li.find('a', class_ = 'sh_blog_title _sp_each_url _sp_each_title')
            title = find_a.text
            url = find_a.get('href')
            desc_get = li.find('dd', class_ = 'sh_blog_passage')
            desc = desc_get.text
            result_list.append({
                    'title' : title,
                    'url' : url,
                    'desc' :desc
            })

    return result_list

page_num = int(input("몇 페이지까지? : "))

topic = input("검색어? : ")

print()
page_crawling = page_crawling(page_num, topic)
print(page_crawling)
len(page_crawling)
```

```python
def get_day_sise(code, page_num) :
    base_url = 'https://finance.naver.com/item/sise_day.nhn?code=005930'
    
    result_list = []
    for page in range(1,page_num+1) :
        params = {
            'code' : code,
            'page' : page
        }
        resp = re.get(base_url, params = params)
        soup = bs(resp.text)
        
        tr_tags = soup.select('table.type2 tr')

        th_tags = tr_tags[0].find_all('th')
        for th in th_tags :
            title_list.append(th.text)
        for tr_tag in tr_tags :
            td_tags = tr_tag.find_all('td')
            data_list = []
            if len(td_tags)<7 :
                continue
            for td in td_tags :
                data_list.append(td.text.strip())
            result_list.append(
                dict(zip(title_list, data_list))
            )
        
    return result_list

sise = get_day_sise('005930',5)
sise
```
